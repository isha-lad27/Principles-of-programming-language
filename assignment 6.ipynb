```python
#1) Write a function to find the maximum of three numbers.
def max_of_three(a, b, c):
    # Compare three numbers and return the largest one
    return max(a, b, c)

# Function call
result = max_of_three(10, 25, 15)
print("Maximum of three numbers:", result)
```

```python
#2) Write a function to check if a number is prime or not.
def is_prime(n):
    if n <= 1:  # 0 and 1 are not prime
        return False
    for i in range(2, int(n**0.5)+1):  # Check divisibility up to square root of n
        if n % i == 0:
            return False
    return True

# Input number from user
num = int(input("Enter a number: "))

# Call the function and print result
if is_prime(num):
    print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

```python
#3) Write a function to calculate the factorial of a number (both iterative and recursive).
def factorial_iterative(n):
    fact = 1
    for i in range(1, n+1):
        fact *= i
    return fact

# Function to calculate factorial recursively
def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial_recursive(n-1)

# Input number from user
num = int(input("Enter a number: "))

# Call iterative function
print("Factorial (iterative):", factorial_iterative(num))

# Call recursive function
print("Factorial (recursive):", factorial_recursive(num))
```

```python
#4) Write a function that accepts a list and returns a new list with unique elements. 
def unique_elements(lst):
    #Empty list to store unique values
    unique_list = []
    
    # Loop through each element in the input list
    for i in lst:
        # Check if the element is not already in unique_list
        if i not in unique_list:
            unique_list.append(i)
    
    # Return the list with unique elements
    return unique_list
#Print result
print(unique_elements([1, 2, 3, 2, 4, 1, 5]))
```

```python
#5) Write a function to calculate the sum of all even numbers in a list.
def sum_of_even(lst):
    # Initialize total sum as 0
    total = 0
    
    # Loop through each element in the list
    for i in lst:
        # Check if the number is even
        if i % 2 == 0:
            total += i
    
    # Return the sum of even numbers
    return total

# Example
print(sum_of_even([1, 2, 3, 4, 6, 7]))
```

```python
#1) Write a function to convert temperature from Celsius to Fahrenheit and vice versa. 
# Function to convert temperature
def convert_temperature(temp, scale):
    if scale == "C":  # Celsius to Fahrenheit
        return (temp * 9/5) + 32
    elif scale == "F":  # Fahrenheit to Celsius
        return (temp - 32) * 5/9
    else:
        return None

# Input temperature from user
temp = float(input("Enter temperature: "))
scale = input("Enter scale (C for Celsius, F for Fahrenheit): ").upper()

# Call the function and print result
converted = convert_temperature(temp, scale)
if converted is not None:
    if scale == "C":
        print(f"{temp}°C = {converted:.2f}°F")
    else:
        print(f"{temp}°F = {converted:.2f}°C")
else:
    print("Invalid scale entered!")
```

```python
#2) Write a function that accepts a string and returns the count of vowels and consonants.
# Function to count vowels and consonants
def count_vowels_consonants(s):
    vowels = "aeiouAEIOU"
    v_count = 0
    c_count = 0
    for char in s:
        if char.isalpha():  # Check if character is a letter
            if char in vowels:
                v_count += 1
            else:
                c_count += 1
    return v_count, c_count

# Input string from user
text = input("Enter a string: ")

# Call the function
vowels, consonants = count_vowels_consonants(text)

# Print result
print("Number of vowels:", vowels)
print("Number of consonants:", consonants)
```

```python
#3) Write a function that accepts variable number of arguments and returns their average
# Function to calculate average of variable number of arguments
def average(*args):
    total = 0
    count = len(args)
    for num in args:
        total += num
    return total / count if count != 0 else 0

# Call the function and input numbers from user
nums = list(map(float, input("Enter numbers separated by space: ").split()))

# Call the function using *nums
print("Average is:", average(*nums))
```

```python
#4) Write a lambda function to find the maximum of three numbers.

maximum = lambda a, b, c: a if (a >= b and a >= c) else (b if b >= c else c)

# Input three numbers from user
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
z = int(input("Enter third number: "))

# Call the lambda function and print result
print("Maximum number is:", maximum(x, y, z))

```

```python
#1) Write a recursive function to calculate the sum of digits of a number.
# Recursive function to calculate sum of digits
def sum_of_digits(n):
    if n == 0:
        return 0
    else:
        return n % 10 + sum_of_digits(n // 10)

# Input number from user
num = int(input("Enter a number: "))

# Call the function and print result
print("Sum of digits:", sum_of_digits(num))
```

```python
#2) Write a function that accepts a list of numbers and returns a dictionary with count of even 
#and odd numbers. 
#Example: [1,2,3,4,5] should return {'even': 2, 'odd': 3}
def count_even_odd(lst):
    result = {'even': 0, 'odd': 0}  # Dictionary to store counts
    for num in lst:
        if num % 2 == 0:
            result['even'] += 1
        else:
            result['odd'] += 1
    return result

# Input list from user
numbers = list(map(int, input("Enter numbers separated by space: ").split()))

# Call the function and print result
print("Even and Odd count:", count_even_odd(numbers))
```
