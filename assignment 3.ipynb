```python
#1) Write a program to find the maximum and minimum element in a tuple without using max() and min() functions.
t = (4, 7, 1, 9, 3)
# Assume first element as both maximum and minimum
maximum = t[0]
minimum = t[0]

# Check each element in the tuple
for i in t:
    if i > maximum:
        maximum = i
    if i < minimum:
        minimum = i
        
# Print results
print("Maximum:", maximum)
print("Minimum:", minimum)
```

```python
#2) Write a program to count the occurrence of each element in a tuple.

#Given tuple (you can also take input if needed)
t = (1, 2, 3, 2, 1, 3, 2)

# Empty dictionary to store count
count_dict = {}

# Count each element in the tuple
for i in t:
    if i in count_dict:
        count_dict[i] += 1
    else:
        count_dict[i] = 1

# Print the result
for key, value in count_dict.items():
    print(f"{key}:{value}")
```

```python
#3) Write a program to convert a tuple of strings into a single string.
# Given tuple of strings
t = ("Hello", "World", "Python")

# Empty string to store result
result = ""

# Join each string from the tuple
for word in t:
    result += word

# Print the result
print(result)
```

```python
#4) Write a program to find the sum of all even numbers in a tuple.
# Given tuple of numbers
t = (1, 2, 3, 4, 5, 6, 8)

# Variable to store sum of even numbers
even_sum = 0


# Check each element in the tuple
for i in t:
    if i % 2 == 0:
        even_sum += i


# Print the result
print("Sum of even numbers:", even_sum)
```

```python
#5) Write a program to remove duplicates from a tuple (return a new tuple with unique elements).
# Given tuple with duplicate values
t = (1, 2, 3, 2, 4, 1, 5)

# Empty list to store unique elements
unique = []

# Check each element in the tuple
for i in t:
    if i not in unique:
        unique.append(i)
        
# Convert list back to tuple
new_tuple = tuple(unique)

# Print the new tuple
print("Tuple without duplicates:", new_tuple)
```

```python
#1) Write a program to find the second largest element in a tuple.

# Define a tuple of numbers
t = (10, 5, 8, 20, 15)

#Initialize largest and second largest with first element
largest = t[0]
second_largest = t[0]

#Loop through each element in the tuple
for i in t:
    if i > largest:
        # Update second largest before updating largest
        second_largest = largest
        largest = i
        # Update second largest if i is between largest and second_largest
    elif i > second_largest and i != largest:
        second_largest = i

# Print the second largest element
print("Second largest element:", second_largest)
```

```python
#2) Write a program to sort a tuple of tuples based on the second element.
#Define a tuple of tuples
t = ((1, 5), (3, 2), (2, 8))

#Sort the tuple based on the second element of each inner tuple
sorted_tuple = tuple(sorted(t, key=lambda x: x[1]))

# Print the sorted tuple
print("Sorted tuple:", sorted_tuple)
```

```python
#3) Write a program to check if all elements in a tuple are unique.
# Given tuple
t = (1, 2, 3, 4, 5)

# Convert tuple to set
s = set(t)

# Compare lengths of tuple and set
if len(t) == len(s):
    print("All elements are unique")
else:
    print("There are duplicates")
```

```python
#4) Write a program to swap the first and last element of a tuple.
# Program to swap first and last element of a tuple

# Given tuple
t = (10, 20, 30, 40, 50)

# Convert tuple to list (tuples are immutable)
lst = list(t)

# Swap first and last elements
lst[0], lst[-1] = lst[-1], lst[0]

# Convert back to tuple
t = tuple(lst)

# Print the result
print("Tuple after swapping:", t)
```

```python
#1) Write a program to find common elements between two tuples.
# Given tuples
t1 = (1, 2, 9, 4, 7)
t2 = (3, 1, 2, 5, 7)

# Empty list to store common elements
common = []

# Check each element of first tuple
for i in t1:
    if i in t2:
        common.append(i)

# Convert list to tuple (optional)
common_tuple = tuple(common)

# Print the result
print("Common elements:", common_tuple)
```

```python
#2) Write a program to create a tuple of squares of numbers from 1 to n.
n = 5

# Empty list to store squares
squares = []

# Calculate square of each number from 1 to n
for i in range(1, n + 1):
    squares.append(i * i)

# Convert the list of squares to a tuple
squares_tuple = tuple(squares)

# Print the tuple of squares
print("Tuple of squares:", squares_tuple)
```
