```python
#2) Write a program which accepts 6 integer values and prints "DUPLICATES" if any of the values entered are duplicates otherwise it prints "ALL UNIQUE".
# Create an empty list to store values
values = []
# Accept 6 integer values from the user
for i in range(6):
    num = int(input(f"Enter integer {i+1}: "))
    values.append(num)
    
# Compare length of list and set to detect duplicates
if len(values) != len(set(values)):
    print("DUPLICATES")
else:
    print("ALL UNIQUE")
```

```python
#4) Write a program which finds sum of digits of a number.
# Accept an integer input
n = int(input("Enter a number: "))

# Variable to store sum of digits
sum_digits = 0

# Loop until the number becomes zero
while n > 0:
     # Add last digit to sum
    sum_digits += n % 10
    # Remove last digit from number
    n //= 10

# Display the sum of digits
print("Sum of digits:", sum_digits)
```

```python
#5) Write a program which prints Fibonacci series of a number.
# Accept number of terms from user
n = int(input("Enter number of terms: "))

# Initialize the first two Fibonacci numbers
a, b = 0, 1

# Print Fibonacci series
print("Fibonacci Series:")

for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

```python
#1) Write a program which accept an integer value 'n' and display all prime numbers till 'n'.
n = int(input("Enter a number: "))

# Print prime numbers up to n
print("Prime numbers up to", n, "are:")

for num in range(2, n + 1):
    is_prime = True
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num, end=" ")
```

```python
#4) Write a program to reverse a given number.
# Accept an integer number from the user
num = int(input("Enter a number: "))

# Variable to store reversed number
reverse_num = 0

# Loop to reverse the number
while num > 0:
    digit = num % 10             # Get last digit
    reverse_num = reverse_num * 10 + digit
    num = num // 10              # Remove last digit

# Print the reversed number
print("Reversed number is:", reverse_num)
```

```python
#1 Write a Sequential search function which searches an item in a sorted list. The function should return the index of element to be searched in the list.
#should return the index of element to be searched in the list.
def sequential_search(arr, key):
    # Traverse the list one by one
    for index in range(len(arr)):
        # Check if current element matches the key
        if arr[index] == key:  
            return index       # Return index if found
    return -1                   # Return -1 if not found

# Example usage
numbers = [10, 20, 30, 40, 50, 60]

# Element to search
search_item = 40

# Call the function
result = sequential_search(numbers, search_item)

# Display result
if result != -1:
    print("Element found at index:", result)
else:
    print("Element not found")
```
