```python
#1) Write a program to create a dictionary with roll number as key and name as value for 5 students. Display all key-value pairs.
# Empty dictionary
students = {}

# Input roll numbers and names for 5 students
for i in range(5):
    roll = input("Enter roll number: ")
    name = input("Enter name: ")
    students[roll] = name

# Display all key-value pairs
print("Student dictionary:", students)
```

```python
#2) Write a program to count the frequency of each word in a given sentence using a dictionary.

# Input sentence from user
sentence = input("Enter a sentence: ")

# Split sentence into words
words = sentence.split()

# Empty dictionary to store word frequency
freq = {}

# Count frequency of each word
for w in words:
    if w in freq:
        freq[w] += 1
    else:
        freq[w] = 1

# Print the dictionary
print("Word frequency:", freq)
```

```python
#3) Write a program to merge two dictionaries entered by the user.

# Input first dictionary from user
dict1 = {}
n1 = int(input("How many elements in first dictionary? "))
for i in range(n1):
    key = input("Enter key: ")
    value = input("Enter value: ")
    dict1[key] = value

# Input second dictionary from user
dict2 = {}
n2 = int(input("How many elements in second dictionary? "))
for i in range(n2):
    key = input("Enter key: ")
    value = input("Enter value: ")
    dict2[key] = value

# Merge dictionaries
dict1.update(dict2)

# Print merged dictionary
print("Merged dictionary:", dict1)
```

```python
#4) Write a program to find the key with the maximum value in a dictionary


# Input dictionary from user
d = {}
n = int(input("Enter number of elements in dictionary: "))
for i in range(n):
    key = input("Enter key: ")
    value = int(input("Enter value: "))
    d[key] = value

# Assume first key as maximum
max_key = list(d.keys())[0]

# Check each key to find maximum value
for key in d:
    if d[key] > d[max_key]:
        max_key = key

# Print the key with maximum value
print("Key with maximum value:", max_key)
```

```python
#5) Write a program to check if a given key exists in a dictionary.
# Program to check if a key exists in a dictionary

# Example dictionary
d = {'a': 10, 'b': 25, 'c': 15}

# Input key to check
key = input("Enter key to check: ")

# Check if key exists
if key in d:
    print(key, "exists in the dictionary")
else:
    print(key, "does not exist in the dictionary")
```

```python
#1) Write a program to remove a key from a dictionary and handle the case when key doesn't exist.

d = {'a': 10, 'b': 25, 'c': 15}

# Input key to remove
key = input("Enter key to remove: ")

# Check if key exists
if key in d:
    d.pop(key)  # Remove key
    print("Updated dictionary:", d)
else:
    print("Key does not exist in the dictionary")
```

```python
#2) Write a program to sort a dictionary by its values in descending order.

# Example dictionary
d = {'a': 10, 'b': 25, 'c': 15}

# Sort dictionary items by value (descending) and convert back to dictionary
sorted_d = dict(sorted(d.items(), key=lambda item: item[1], reverse=True))

# Print sorted dictionary
print("Dictionary sorted by values (descending):", sorted_d)

```

```python
#3) Write a program to create a dictionary from two lists - one containing keys and another containing values
# Program to create a dictionary from two lists

# Input list of keys
keys = input("Enter keys separated by space: ").split()

# Input list of values
values = input("Enter values separated by space: ").split()

# Create dictionary by pairing keys and values
d = {}
for i in range(len(keys)):
    d[keys[i]] = values[i]

# Print the dictionary
print("Dictionary:", d)
```

```python
#1) Write a program to find all keys with the same value in a dictionary

# Example dictionary
d = {'a': 10, 'b': 20, 'c': 10, 'd': 30, 'e': 20}

# Input value to search for
value = int(input("Enter the value to search for: "))

# List to store keys with the given value
keys_with_value = []

# Check each key-value pair
for key in d:
    if d[key] == value:
        keys_with_value.append(key)

# Print the result
if keys_with_value:
    print("Keys with value", value, ":", keys_with_value)
else:
    print("No key found with value", value)
```

```python
#2) Write a program to create a dictionary where keys are numbers from 1 to n and values are their cubes


# Input n from user
n = int(input("Enter a number: "))

# Empty dictionary
d = {}

# Loop from 1 to n and store cubes
for i in range(1, n+1):
    d[i] = i**3

# Print the dictionary
print("Dictionary of cubes:", d)
```
