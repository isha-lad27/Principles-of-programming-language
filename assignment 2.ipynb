<h1>Question Set A<h1>

```python
#1) Write a program to count the number of vowels in a given string.
# Accept a string from the user
text = input("Enter a string: ")

# Define vowels
vowels = "aeiouAEIOU"

# Initialize vowel count
count = 0

#Check each character in the string
for ch in text:
    if ch in vowels:
        count += 1
        
# Print the result
print("Number of vowels:", count)
```

```python
#2) Write a program to check if a string is a palindrome (reads the same forwards and backwards).
# Accept a string from the user
text = input("Enter a string: ")

# Reverse the string
reverse_text = text[::-1]

# Check if original string and reversed string are same
if text == reverse_text:
    print("Palindrome")
else:
    print("Not a Palindrome")
```

```python
#3) Write a program to reverse a string without using built-in reverse functions.
# Accept a string from the user
text = input("Enter a string: ")

# Variable to store reversed string
reversed_str = ""

# Loop through the string from end to start
for ch in text:
    reversed_str = ch + reversed_str
    
# Print the reversed string
print("Reversed string:", reversed_str)
```

```python
#4) Write a program to count the frequency of each character in a string.
# Accept a string from the user
text = input("Enter a string: ")

# Dictionary to store character frequencies
frequency = {}

# Loop through each character in the string
for ch in text:
    if ch in frequency:
        frequency[ch] += 1         # Increase count if character exists
    else:
        frequency[ch] = 1          # Initialize count for new character
# Print the character frequencies
for ch, count in frequency.items():
    print(f"{ch}:{count}")
```

```python
#5) Write a program to remove all whitespace from a string.
# Accept a string from the user
text = input("Enter a string: ")
# String to store result without spaces
result = ""

# Loop through each character
for ch in text:          
    if ch != " ":           # Check for space
        result += ch        # Add non-space characters

# Print the result
print("String without whitespace:", result)
```

```python
#1) Write a program to find the first non-repeating character in a string.
# Accept a string from the user
text = input("Enter a string: ")

# Loop through each character in the string
for ch in text:
    # Check if the character appears only once in the string
    if text.count(ch) == 1:
        # Print the first non-repeating character
        print("First non-repeating character:", ch)
        break
else:
    print("No non-repeating character found")
```

```python
#2) Write a program to check if two strings are anagrams (contain same characters in different order).
# Program to check whether two strings are anagrams
str1 = input("Enter first string: ")
str2 = input("Enter second string: ")

# Check if both strings are anagrams
if sorted(str1) == sorted(str2):
    print("Anagrams")
else:
    print("Not anagrams")
```

```python
#3) Write a program to convert a string to title case without using the title() method.
text = input("Enter a string: ")
words = text.split()
result = ""

# Convert first letter of each word to uppercase
for word in words:
    result += word[0].upper() + word[1:] + " "

print("Title Case:", result.strip())
```

```python
#4) Write a program to replace all occurrences of a substring with another substring in a given string.
text = input("Enter the main string: ")
old = input("Enter substring to replace: ")
new = input("Enter new substring: ")

result = text.replace(old, new)
print("Updated string:", result)
```

```python
#1) Write a program to find the longest word in a sentence.
# Take sentence from user
sentence = input("Enter a sentence: ")

# Split the sentence into words
words = sentence.split()

# Assume first word is the longest
longest = words[0]

# Compare length of each word
for word in words:
    if len(word) > len(longest):
        longest = word
        
#Print the longest word
print("Longest word:", longest)
```

```python
#2) Write a program to check if a string contains only digits, only alphabets, or both.
# Take input from the user
text = input("Enter a string: ")

# Flags to check alphabets and digits
has_alpha = False
has_digit = False

# Check each character in the string
for ch in text:
    if ch.isalpha():
        has_alpha = True
    elif ch.isdigit():
        has_digit = True
        
# Print result based on flags
if has_alpha and has_digit:
    print("String contains both alphabets and digits")
elif has_alpha:
    print("String contains only alphabets")
elif has_digit:
    print("String contains only digits")
else:
    print("String contains neither alphabets nor digits")
```
