```python
#1) Create a class called Rectangle with attributes length and width. Add methods to calculate area and perimeter.
class Rectangle:

    # Constructor to initialize values
    def __init__(self, length, width):
        self.length = length
        self.width = width

    # Method to calculate area
    def area(self):
        return self.length * self.width

    # Method to calculate perimeter
    def perimeter(self):
        return 2 * (self.length + self.width)

# Input from user
l = int(input("Enter length: "))
w = int(input("Enter width: "))

# Create object of class
r = Rectangle(l, w)

# Display results
print("Area of Rectangle:", r.area())
print("Perimeter of Rectangle:", r.perimeter())
```

```python
#2) Create a class BankAccount with methods to deposit, withdraw, and display balance.
class BankAccount:

    def __init__(self, balance):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        self.balance -= amount

    def show(self):
        print("Balance:", self.balance)

# Input
b = int(input("Enter balance: "))
a = BankAccount(b)

# Deposit
d = int(input("Enter deposit amount: "))
a.deposit(d)

# Withdraw
w = int(input("Enter withdraw amount: "))
a.withdraw(w)

# Display balance
a.show()
```

```python
#3) Create a parent class Vehicle and child classes Car and Bike that inherit from it. Override a method in child classes.
# Parent class
class Vehicle:
    def move(self):
        print("Vehicle is moving")

# Child class Car
class Car(Vehicle):
    def move(self):   # Overriding method
        print("Car is moving on road")

# Child class Bike
class Bike(Vehicle):
    def move(self):   # Overriding method
        print("Bike is moving on road")

# Create objects
v = Vehicle()
c = Car()
b = Bike()

# Call methods
v.move()
c.move()
b.move()
```

```python
#4) Create a class Student with private attributes and public methods to access them.
class Student:

    # Constructor
    def __init__(self, name, marks):
        self.__name = name      # private attribute
        self.__marks = marks    # private attribute

    # Public method to access name
    def get_name(self):
        return self.__name

    # Public method to access marks
    def get_marks(self):
        return self.__marks

# Input from user
n = input("Enter student name: ")
m = int(input("Enter marks: "))

# Create object
s = Student(n, m)

# Access private data using public methods
print("Name:", s.get_name())
print("Marks:", s.get_marks())
```

```python
#5) Create a class Circle with method to calculate area. Overload the + operator to add two circles (add their radii).
class Circle:

    # Constructor
    def __init__(self, radius):
        self.radius = radius

    # Method to calculate area
    def area(self):
        return 3.14 * self.radius * self.radius

    # Overloading + operator to add radii of two circles
    def __add__(self, other):
        return Circle(self.radius + other.radius)

# Input radii
r1 = int(input("Enter radius of first circle: "))
r2 = int(input("Enter radius of second circle: "))

# Create objects
c1 = Circle(r1)
c2 = Circle(r2)

# Add two circles
c3 = c1 + c2

# Display results
print("Area of first circle:", c1.area())
print("Area of second circle:", c2.area())
print("Radius after adding circles:", c3.radius)
print("Area of new circle:", c3.area())
```

```python
#1) Create a class hierarchy: Animal -> Mammal -> Dog. Demonstrate multilevel inheritance.
class Animal:
    def eat(self):
        print("Animal eats food")

# Child class of Animal
class Mammal(Animal):
    def walk(self):
        print("Mammal walks on land")

# Child class of Mammal
class Dog(Mammal):
    def bark(self):
        print("Dog barks")

# Create object of Dog class
d = Dog()

# Call methods
d.eat()
d.walk()
d.bark()
```

```python
#3) Create a class Book with special methods __str__, __len__, and __eq__.
class Book:

    def __init__(self, title, pages):
        self.title = title
        self.pages = pages

    # String representation
    def __str__(self):
        return "Book Title: " + self.title

    # Length of book (number of pages)
    def __len__(self):
        return self.pages

    # Compare two books
    def __eq__(self, other):
        return self.pages == other.pages

# Create objects
b1 = Book("Python", 200)
b2 = Book("Java", 200)

# Use special methods
print(b1)                # calls __str__
print("Pages:", len(b1)) # calls __len__
print("Books equal:", b1 == b2)  # calls __eq__
```

```python
#4) Create a class Temperature with property decorators for Celsius and Fahrenheit conversion.
class Temperature:

    def __init__(self, celsius):
        self._celsius = celsius

    # Property to get Celsius
    @property
    def celsius(self):
        return self._celsius

    # Property to get Fahrenheit
    @property
    def fahrenheit(self):
        return (self._celsius * 9/5) + 32

# Create object
t = Temperature(25)

# Display values
print("Celsius:", t.celsius)
print("Fahrenheit:", t.fahrenheit)
```

```python
#1) Create an abstract class Shape with abstract method area(). Create subclasses Rectangle,Circle, and Triangle that implement the area method.
from abc import ABC, abstractmethod
# Abstract class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# Rectangle class
class Rectangle(Shape):
    def area(self):
        return 10 * 5

# Circle class
class Circle(Shape):
    def area(self):
        return 3.14 * 3 * 3

# Triangle class
class Triangle(Shape):
    def area(self):
        return 0.5 * 6 * 4

# Objects
r = Rectangle()
c = Circle()
t = Triangle()

# Display area
print("Rectangle Area:", r.area())
print("Circle Area:", c.area())
print("Triangle Area:", t.area())
```

```python
#2) Create a class Library that manages a collection of books. Implement methods to add books, remove books, search books, and display all books.
# Class Library
class Library:

    def __init__(self):
        self.books = []

    # Add book
    def add_book(self, book):
        self.books.append(book)

    # Remove book
    def remove_book(self, book):
        if book in self.books:
            self.books.remove(book)
        else:
            print("Book not found")

    # Search book
    def search_book(self, book):
        if book in self.books:
            print("Book found")
        else:
            print("Book not found")

    # Display all books
    def show_books(self):
        print("Books in library:", self.books)

# Create object
lib = Library()
# Add books
lib.add_book("Python")
lib.add_book("Java")
lib.add_book("C")

# Search book
lib.search_book("Python")

# Remove book
lib.remove_book("Java")

# Display books
lib.show_books()
```
