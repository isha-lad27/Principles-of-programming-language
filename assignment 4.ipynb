```python
#1) Write a program to find the union, intersection, and difference of two sets entered by the user.


# Take input from user for first set
set1 = set(map(int, input("Enter elements of first set separated by space: ").split()))

# Take input from user for second set
set2 = set(map(int, input("Enter elements of second set separated by space: ").split()))

# Union of sets
union_set = set1 | set2  # or set1.union(set2)

# Intersection of sets
intersection_set = set1 & set2  # or set1.intersection(set2)

# Difference of sets (set1 - set2)
difference_set = set1 - set2  # or set1.difference(set2)

# Print results
print("Union:", union_set)
print("Intersection:", intersection_set)
print("Difference (set1 - set2):", difference_set)
```

```python
#2) Write a program to check if a set is a subset of another set. 
# Program to check if one set is a subset of another

# Input first set
set1 = set(map(int, input("Enter elements of first set: ").split()))

# Input second set
set2 = set(map(int, input("Enter elements of second set: ").split()))

# Check if set1 is subset of set2
if set1.issubset(set2):
    print("First set is a subset of second set")
else:
    print("First set is not a subset of second set")
```

```python
#3) Program to remove duplicates from a list using sets
#Original list with duplicate elements
lst = [1, 2, 3, 2, 4, 1, 5]

#Convert list to set to remove duplicates, then back to list
unique_list = list(set(lst))


# Print the list after removing duplicates
print("List after removing duplicates:", unique_list)
```

```python
#4) Program to find elements present in the first set but not in the second set
# Input first set from user
s1 = set(input("Enter elements of first set: ").split())

# Input second set from user
s2 = set(input("Enter elements of second set: ").split())

# Find difference 
result = s1 - s2

# Print result
print("Elements in first set but not in second:", result)
```

```python
#5) Write a program to check if two sets are disjoint (have no common elements).
# Input first set from user
s1 = set(input("Enter elements of first set: ").split())

# Input second set from user
s2 = set(input("Enter elements of second set: ").split())

# Check if sets have no common elements
if s1.isdisjoint(s2):
    print("The sets are disjoint")
else:
    print("The sets are not disjoint")
```

```python
#1) Write a program to find symmetric difference between two sets.
# Input first set from user
s1 = set(input("Enter elements of first set: ").split())
# Input second set from user
s2 = set(input("Enter elements of second set: ").split())

# Symmetric difference (elements in set1 or set2 but not in both)
sym_diff = s1 ^ s2

# Print result
print("Symmetric Difference:", sym_diff)
```

```python
#2) Write a program to add multiple elements to a set at once from user input.
#Create an empty set
s = set()

#Take input elements (space-separated) from the user
elements = input("Enter elements to add to the set: ").split()

#Add all input elements to the set using update()
s.update(elements)

# Print the updated set
print("Updated set:", s)
```

```python
#3) Write a program to find the length of a set and check if a specific element exists in it.

# Input set from user
s = set(input("Enter elements of the set: ").split())

# Input element to check
element = input("Enter element to check: ")

#length of the set
print("Length of the set:", len(s))

#Check if element exists in the set
if element in s:
    print(f"{element} exists in the set")
else:
    print(f"{element} does not exist in the set")
```

```python
#1) Write a program that takes two sentences as input and finds: 
#● Common words in both sentences 
#● Words unique to the first sentence 
#● Words unique to the second sentence 

# Input two sentences from user
sentence1 = input("Enter first sentence: ")
sentence2 = input("Enter second sentence: ")

# Split sentence into words 
set1 = set(sentence1.split())
set2 = set(sentence2.split())

# Common words
common_words = set1 & set2

# Words unique to first sentence
unique_first = set1 - set2

# Words unique to second sentence
unique_second = set2 - set1

# Print results
print("Common words:", common_words)
print("Words unique to first sentence:", unique_first)
print("Words unique to second sentence:", unique_second)
```

```python
#2) Write a program to perform all set operations (union, intersection, difference, symmetric difference) on three sets. 
s1 = set(input("Enter elements of first set: ").split())
s2 = set(input("Enter elements of second set: ").split())
s3 = set(input("Enter elements of third set: ").split())

# Union
union_set = s1 | s2 | s3

# Intersection
intersection_set = s1 & s2 & s3

# Difference
difference_set = s1 - s2 - s3

# Symmetric Difference
sym_diff = s1 ^ s2 ^ s3

# Print results
print("Union:", union_set)
print("Intersection:", intersection_set)
print("Difference (S1 - S2 - S3):", difference_set)
print("Symmetric Difference:", sym_diff)
```
